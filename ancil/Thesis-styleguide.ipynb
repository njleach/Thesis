# Default axes style

The document \textwidth is 5.848611111111111 inches.

```python
fig,ax = plt.subplots()

# title
ax.text(0.02,0.98,'Title',transform=ax.transAxes,va='top',ha='left',fontweight='bold')

# axes labels
ax.set_xlabel('Xlabel (unit)')
ax.set_ylabel('Ylabel (unit)')

# legend
for i in range(5):
    plt.plot([],[],label='handle{i}'.format(i=str(i)))
    
## if multi-panel or no space in plot
l1 = plt.legend(frameon=False,loc='center left',bbox_to_anchor=(1,0.5),title='to side',fontsize='small')
ax.add_artist(l1)
## else
plt.legend(frameon=False,title='in axes',fontsize='small')

sn.despine()
fig.patch.set_facecolor((1,1,1))
fig.dpi=120
```
