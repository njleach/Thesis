# Default axes style

```python
fig,ax = plt.subplots()

# title
ax.text(0.02,0.98,'Title',transform=ax.transAxes,va='top',ha='left',fontweight='bold')

# axes labels
ax.set_xlabel('xlabel / unit(s)')
ax.set_ylabel('ylabel / unit(s)')

# legend
for i in range(5):
    plt.plot([],[],label='handle{i}'.format(i=str(i)))
    
## if multi-panel or no space in plot
l1 = plt.legend(frameon=False,loc='center left',bbox_to_anchor=(1,0.5),title='to side')
ax.add_artist(l1)
## else
plt.legend(frameon=False,title='in axes')

sn.despine()
fig.patch.set_facecolor((1,1,1))
fig.dpi=120
```
